```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F_xn(x, n):
  if (n < 0):
    return x**(1/(-1 * n))
  else:
    return x**n
```

```python
N = 1000
x = np.linspace(0, 1, N)
```

```python
plt.figure(figsize=(7, 6))
for i in range(-10, 10):
  if i == 0:
    continue
  plt.plot(x, F_xn(x, i), label=f'x^{i}')
plt.xlabel('x')
plt.ylabel('y')
plt.margins(0.05)
plt.show()
```
