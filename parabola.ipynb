```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
  return x**2
```

```python
def F_dx(x):
  h = 1e-6
  return (F(x + h) - F(x)) / h
```

```python
N = 30
x = np.linspace(-20, 20, N)
y = F(x)
dy = F_dx(x)
```

```python
plt.plot(x, y, label=r'$f(x) = x^2$')
plt.plot(x, dy, label=r'$f\'(x) = 2x$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
