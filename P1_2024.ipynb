## Questão 2 (3.5 pontos)

O mapa logístico é uma função matemática discreta que descreve a dinâmica de populações em crescimento. A fórmula de recorrência do mapa logístico é dada por:
$$x_{n+1} = r x_n  (1 - x_n)$$

onde:

$x_n$ é a população na geração $n$ ($0 \le x_n \le 1$)

$x_{n+1}$ é a população na próxima geração

$r$ é o parâmetro de crescimento ($0 \le r \le 4$).


Escreva um programa em Python que simule o crescimento populacional utilizando o mapa logístico. O programa deve receber como entrada o valor inicial da população $x_0$, o parâmetro de crescimento $r$ e o número de gerações $n$. Em seguida, ele deve calcular e exibir a população em cada geração, até a geração $n$. Teste seu programa considerando: $x_0 = 0.4, r = 3.0, n = 10$.


```python
# Altere 'valDefault' para false para inserir parâmetros personalizados
valDefault = True;

def MapaLogistico(x0, n, r):
  Xn = []
  Xn.append(x0)

  for i in range(n+1):
    Xn.append(r * Xn[i] * (1 - Xn[i]))
    print(f"Geraçao {i}: {Xn[i]}")


def main():
  if valDefault:
    x0 = 0.4
    r = 3.0
    n = 10
  else:
    x0 = float(input())
    r = float(input())
    n = int(input())

  if (r < 0 or r > 4):
    print("Parâmetro r: iválido")
    return

  MapaLogistico(x0, n, r)

main()
```

O fluxograma abaixo representa um algoritmo para obter uma aproximação do valor da raiz quadrada de um número N através do método babilônico. Codifique o algoritmo abaixo em uma função Python que recebe como entrada o valor de N e do nível de acurácia desejado. Teste seu algoritmo para a N = 80 e acuracia = 0.00001. Compare seu resultado com o valor de $\sqrt{80}$ obtido com a função para raiz quadrada do NumPy.
Flowchart.png


```python
import numpy as np

def Babilonico(N, acuracia):
  x = N
  y = 1
  while (True):
    if (x-y) > acuracia:
      x = (x + y) / 2
      y = N / x
    else:
      print(f"Valor de X utilizando Método Babilônico: {x}")
      print(f"Valor raiz quadrada de {N} com a biblioteca NumPy: {np.sqrt(N)}")
      break;

def main():
  Babilonico(80, 0.00001)

main()
```

## Questão 4 (2.5 pontos)

Produza uma figura contendo um grid de 4 gráficos (2 linhas e 2 colunas), utilizando subplots, das seguintes funções $f(x)$ para $x$ no intervalo [-2,2]:

$f(x) = cos(x)$

$f(x) = sen(x)$

$f(x) = x^2 +  \frac{x}{2} -1$

$f(x) = 5 x + 3$

```python
import numpy as np
import math
import matplotlib.pyplot as plt

interval = np.arange(-2, 2, 0.1)

def Function3():
    ret = []
    for i in interval:
      ret.append(i**2 + (i/2) - 1)
    return ret

def Function4():
    ret = []
    for i in interval:
      ret.append(5 * i + 3)
    return ret

def main():
  plt.plot(interval,np.cos(interval), 'r')
  plt.plot(interval,np.sin(interval), 'b')
  plt.plot(interval,Function3(), 'g')
  plt.plot(interval,Function4(), 'y')
  plt.xlabel('Intervalo')
  plt.ylabel('Resultado')
  plt.title('Gráfico Questão 4')
  plt.legend(['cos (x)', 'sen (x)', 'x² + (x/2) - 1', '5x + 3'])
  plt.savefig('Questao4.png')
  plt.show()

main()
```
