```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import bisect

# Constantes Dadas
v0 = 10
x = 4.6
y = 3.05
g = 9.8

# Minha altura
y0 = 1.80

theta = np.arange(1, np.pi/2, 0.01)

def f(theta):
  term = np.sqrt(v0**2 * np.sin(theta)**2 - 2 * g * (y - y0))
  return x - ((v0 * np.cos(theta) / g) * (v0 * np.sin(theta) + term))

# Método da Bisseção
theta_zero = bisect(f, 1, np.pi/2)
print(f"O ângulo θ onde f(θ) = 0 é: {theta_zero} rad")

# Calcula o tempo para a bola atingir a distância da cesta
t_impact = x / (v0 * np.cos(theta_zero))

# Equações da trajetória
t = np.linspace(0, t_impact, num=500)
x_t = v0 * np.cos(theta_zero) * t
y_t = y0 + v0 * np.sin(theta_zero) * t - 0.5 * g * t**2

# Gráficos
figura = plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.plot(theta, f(theta), color='g', linewidth=2)
plt.title("F(θ)")
plt.ylabel("F(θ)")
plt.xlabel("θ")
plt.grid(True)
plt.subplot(122)
plt.plot(x_t, y_t, color='b', linewidth=2)
plt.title("Trajetória da Bola")
plt.ylabel("Y(t)")
plt.xlabel("X(t)")
plt.axhline(y=y, color='r', linestyle='--', label='Altura da cesta (3.05 m)')
plt.axvline(x=x, color='r', linestyle='--', label='Distância até a cesta (4.6 m)')
plt.scatter(0, y0, color='b', zorder=5, label='Altura do lançamento (1.80 m)')
plt.legend()
plt.tight_layout()
plt.show()
```
