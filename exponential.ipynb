```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
  return np.exp(x)
```

```python
N = 30
x = np.linspace(-2, 10, N)
y = F(x)
```

```python
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('F(X)')
plt.title('e^x')
plt.axvline(x=0, color='gray')
plt.axhline(y=0, color='gray')
plt.show()
```
