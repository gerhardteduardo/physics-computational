```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def F(x):
  return np.exp(x)
```

```python
def F_dx(x):
  h = 1e-6
  return (F(x + h) - F(x)) / h
```

```python
N = 30
x = np.linspace(-2, 10, N)
y = F(x)
dy = F_dx(x)
```

```python
plt.scatter(x, y, label=r'$f(x) = e^x$')
plt.plot(x, dy, label=r'$f\'(x) = e^x$')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(x=0, color='gray')
plt.axhline(y=0, color='gray')
plt.show()
```
