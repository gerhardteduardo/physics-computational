```python
import numpy as np
import matplotlib.pyplot as plt

def solve_schrodinger_hydrogen_L0(N=2000, r_max=100.0):
    """
    Resolve a equação de Schrödinger radial independente do tempo para o
    átomo de hidrogênio com L=0 usando o método das diferenças finitas.

    Args:
        N (int): Número de pontos na grade.
        r_max (float): Raio máximo para a grade.

    Returns:
        tuple: Uma tupla contendo:
            - eigenvalues (np.ndarray): As 10 primeiras energias de estados ligados.
            - eigenvectors (np.ndarray): As funções de onda radiais correspondentes.
            - r (np.ndarray): A grade radial.
    """
    # Define a grade radial
    r = np.linspace(1e-6, r_max, N)
    dr = r[1] - r[0]

    # Potencial para o átomo de hidrogênio (em unidades atômicas)
    V = -1.0 / r

    # Constrói a matriz Hamiltoniana
    # Termo de energia cinética (usando uma diferença finita de três pontos para a segunda derivada)
    T = -0.5 * (-2.0 * np.diag(np.ones(N)) +
                   np.diag(np.ones(N - 1), 1) +
                   np.diag(np.ones(N - 1), -1)) / dr**2

    # Termo de energia potencial
    U = np.diag(V)

    # Hamiltoniana total
    H = T + U

    # Resolve o problema de autovalores
    eigenvalues, eigenvectors = np.linalg.eigh(H)

    # Os autovetores são as funções de onda radiais u(r) = r*R(r)
    # Normaliza as funções de onda
    for i in range(len(eigenvalues)):
        norm = np.sqrt(np.sum(eigenvectors[:, i]**2) * dr)
        eigenvectors[:, i] /= norm

    # Filtra os 10 primeiros estados ligados (energias negativas)
    bound_indices = np.where(eigenvalues < 0)[0][:10]
    bound_energies = eigenvalues[bound_indices]
    bound_wavefunctions = eigenvectors[:, bound_indices]

    return bound_energies, bound_wavefunctions, r

# Resolve a equação
energies, wavefunctions, r = solve_schrodinger_hydrogen_L0()

# --- Imprime as Energias ---
print("Energias de Ligação Calculadas (L=0) em unidades atômicas (Hartrees):")
print("-" * 65)
print("{:<10s} {:<25s} {:<25s}".format("n", "Energia Numérica (Eh)", "Energia Analítica (Eh)"))
print("-" * 65)
for i, energy in enumerate(energies):
    n = i + 1
    analytical_energy = -0.5 / (n**2)
    print("{:<10d} {:<25.6f} {:<25.6f}".format(n, energy, analytical_energy))
print("-" * 65)


# --- Plota as Funções de Onda ---
plt.figure(figsize=(10, 8))
for i in range(4): # Plota as 4 primeiras funções de onda (1s, 2s, 3s, 4s)
    n = i + 1
    # O solver nos dá u(r) = r*R(r). Para obter R(r), dividimos por r.
    # Também ajustamos o sinal para a plotagem convencional.
    radial_wavefunction = wavefunctions[:, i] / r
    if radial_wavefunction[10] < 0: # Heurística para combinar com o sinal convencional
        radial_wavefunction *= -1
    plt.plot(r, radial_wavefunction, label=f'n={n}, L=0 ({n}s)')

plt.title('Funções de Onda Radiais ($R(r)$) para o Átomo de Hidrogênio (L=0)', fontsize=16)
plt.xlabel('Raio (raios de Bohr)', fontsize=12)
plt.ylabel('Amplitude da Função de Onda Radial, $R(r)$', fontsize=12)
plt.xlim(0, 50)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(0, color='black', linewidth=0.8)
plt.legend()
plt.savefig("hydrogen_wavefunctions_L0.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def solve_schrodinger_hydrogen_L0(N=100, r_max=200.0): # Parâmetros refinados
    """
    Resolve a equação de Schrödinger radial independente do tempo para o
    átomo de hidrogênio com L=0 usando o método das diferenças finitas.

    Args:
        N (int): Número de pontos na grade.
        r_max (float): Raio máximo para a grade.

    Returns:
        tuple: Uma tupla contendo:
            - eigenvalues (np.ndarray): As 10 primeiras energias de estados ligados.
            - eigenvectors (np.ndarray): As funções de onda radiais correspondentes.
            - r (np.ndarray): A grade radial.
    """
    # Define a grade radial. A grade vai de dr a r_max, com N pontos.
    # A condição u(0)=0 é imposta pela natureza do problema de autovalores na "caixa".
    r = np.linspace(1e-6, r_max, N)
    dr = r[1] - r[0]

    # Potencial para o átomo de hidrogênio (em unidades atômicas)
    V = -1.0 / r

    # Constrói a matriz Hamiltoniana
    # Termo de energia cinética (usando uma diferença finita de três pontos para a segunda derivada)
    T = -0.5 * (-2.0 * np.diag(np.ones(N)) +
                   np.diag(np.ones(N - 1), 1) +
                   np.diag(np.ones(N - 1), -1)) / dr**2

    # Termo de energia potencial
    U = np.diag(V)

    # Hamiltoniana total
    H = T + U

    # Resolve o problema de autovalores
    eigenvalues, eigenvectors = np.linalg.eigh(H)

    # Normaliza as funções de onda u(r)
    for i in range(len(eigenvalues)):
        norm = np.sqrt(np.sum(eigenvectors[:, i]**2) * dr)
        eigenvectors[:, i] /= norm

    # Filtra os 10 primeiros estados ligados (energias negativas)
    bound_indices = np.where(eigenvalues < 0)[0][:10]
    bound_energies = eigenvalues[bound_indices]
    bound_wavefunctions = eigenvectors[:, bound_indices]

    return bound_energies, bound_wavefunctions, r

# Resolve a equação com os novos parâmetros
energies, wavefunctions, r = solve_schrodinger_hydrogen_L0()

# --- Imprime as Energias (Resultados Mais Precisos) ---
print("Energias de Ligação Calculadas (L=0) em unidades atômicas (Hartrees) - Refinado:")
print("-" * 70)
print("{:<10s} {:<25s} {:<25s}".format("n", "Energia Numérica (Eh)", "Energia Analítica (Eh)"))
print("-" * 70)
for i, energy in enumerate(energies):
    n = i + 1
    analytical_energy = -0.5 / (n**2)
    print("{:<10d} {:<25.6f} {:<25.6f}".format(n, energy, analytical_energy))
print("-" * 70)


# --- Plota as Funções de Onda (Gráfico Mais Suave) ---
plt.figure(figsize=(12, 8))
for i in range(4): # Plota as 4 primeiras funções de onda (1s, 2s, 3s, 4s)
    n = i + 1
    # A solução numérica nos dá u(r). Para obter R(r), dividimos por r.
    # O sinal do autovetor é arbitrário, então ajustamos para a convenção padrão.
    radial_wavefunction = wavefunctions[:, i] / r
    if radial_wavefunction[10] < 0: # Heurística para combinar com o sinal convencional
        radial_wavefunction *= -1
    plt.plot(r, radial_wavefunction, label=f'n={n}, L=0 ({n}s)')

plt.title('Funções de Onda Radiais ($R(r)$) para o Átomo de Hidrogênio (L=0) - Refinado', fontsize=16)
plt.xlabel('Raio (raios de Bohr)', fontsize=12)
plt.ylabel('Amplitude da Função de Onda Radial, $R(r)$', fontsize=12)
plt.xlim(0, 50) # Mantém o zoom para ver os detalhes perto do núcleo
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(0, color='black', linewidth=0.8)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# A equação de Schrödinger radial para L=0 pode ser escrita como um sistema de 2 EDOs de 1ª ordem.
# Seja y[0] = u(r) e y[1] = u'(r) = du/dr.
# Então y[0]' = y[1]
# E y[1]' = u''(r) = -2 * (E - V(r)) * u(r) = -2 * (E + 1/r) * y[0]

def schrodinger_system(r, y, E):
    """ Define o sistema de EDOs para a equação de Schrödinger. """
    # Evita divisão por zero em r=0
    if r < 1e-6:
        r = 1e-6
    return [y[1], -2 * (E + 1.0/r) * y[0]]

# Condições iniciais: u(0) = 0 e u'(0) pode ser qualquer valor pequeno (define a normalização).
# u(r) ~ r para r pequeno, então u'(0) ~ 1.
y0 = [0.0, 1.0]
r_span = [1e-6, 15] # Intervalo de integração

# Energias para testar
E_too_low = -0.52   # Energia muito baixa
E_just_right = -0.5 # Energia correta (estado fundamental)
E_too_high = -0.48  # Energia muito alta

# Resolvendo para cada energia
sol_low = solve_ivp(schrodinger_system, r_span, y0, args=(E_too_low,), dense_output=True, rtol=1e-8)
sol_right = solve_ivp(schrodinger_system, r_span, y0, args=(E_just_right,), dense_output=True, rtol=1e-8)
sol_high = solve_ivp(schrodinger_system, r_span, y0, args=(E_too_high,), dense_output=True, rtol=1e-8)

# Pontos para o gráfico
r_plot = np.linspace(r_span[0], r_span[1], 1000)
u_low = sol_low.sol(r_plot)[0]
u_right = sol_right.sol(r_plot)[0]
u_high = sol_high.sol(r_plot)[0]

# --- Gráfico de Demonstração ---
plt.figure(figsize=(12, 8))
plt.plot(r_plot, u_low, label=f'Energia = {E_too_low} Ha (explode para +∞)', linestyle='--')
plt.plot(r_plot, u_high, label=f'Energia = {E_too_high} Ha (explode para -∞)', linestyle=':')
plt.plot(r_plot, u_right, label=f'Energia = {E_just_right} Ha (solução correta, tende a 0)', linewidth=2.5, color='green')

plt.title('Demonstração do Comportamento da Função de Onda', fontsize=16)
plt.xlabel('Raio (raios de Bohr)', fontsize=12)
plt.ylabel('Função de Onda u(r) = rR(r)', fontsize=12)
plt.axhline(0, color='black', linewidth=0.8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=11)
plt.ylim(-2, 2) # Zoom para ver o comportamento claramente
plt.show()
```
